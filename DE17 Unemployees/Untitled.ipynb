# Generating a sine vs cosine curve

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xrange = np.arange(0, 2*np.pi,0.01)
# cos, sin = [], []
# for number in xrange:
#     cos.append(np.cos(i))
#     sin.append(np.sin(i))
cos = np.cos(xrange)
sin = np.sin(xrange)
```

# Rock, Paper, Scissors Game

```python
import random as rdm
```

```python
choices = ["rock", "paper", "scissor"]
score = {"p1":0, "p2":0}

def wins(p1,p2):
    """ conditions of winning"""
    if p1 == p2:
        pass
    elif p1 == "rock" and p2 == "paper":
        score["p2"] += 1
    elif p1 == "paper" and p2 =="scissor":
        score["p2"] += 1
    elif p1 == "scissor" and p2 == "rock":
        score["p2"] += 1
    else:
        score["p1"] += 1
    
def ezChoice(p1):
    """user input feature. They can reply with numbers because it's easier than typing the word"""
    try:
        p1 = int(p1)
        if p1 == 1:
            p1 = "rock"
        elif p1 == 2:
            p1 = "paper"
        elif p1 == 3:
            p1 = "scissor"
        return p1
    except:
        return None
    
while max(score.values()) < 3:
#     p1 = rdm.choice(choices)
    p1 = input("Choose from rock, paper, scissor: ").lower().strip() 
    p1 = ezChoice(p1)
    while p1 not in choices:
        print("Wrong input")
        p1 = input("Choose from rock(1), paper(2), scissor(3): ").lower().strip()
        p1 = ezChoice(p1)
    p2 = rdm.choice(choices)
    if p1 == "break":
        break
    try:
        print(" p1:",p1,"\n","p2:",p2)
        wins(p1,p2)
    except:
        print(p1,p2, score)
        break
print(score)    
```

```python
# download dictionary and choose a random word
language = "en"
app_id = "fa64e2a8"
app_key = "d2613459e8aca166b1040ae81fa830ef"

def getword(word_id):
    url =  'https://od-api.oxforddictionaries.com:443/api/v2/entries/'  + language + '/'  + word_id.lower()
    return requests.get(url, headers = {'app_id' : app_id, 'app_key' : app_key})

def getwordlist(**kwargs):
    """kwargs (for multiple value entries, separate with comma within string):
    q - The word or inflection to retrieve. This parameter is NOT case-sensitive. 
    fields - 'definitions', 'domains', 'etymologies', 'examples', 'pronunciations', 'regions', 'registers' and 'variantForms'
    """
    url = 'https://od-api.oxforddictionaries.com:443/api/v2/words/'  + language 
    for key,value in kwargs.items():
        url += "&" + key + "=" + value
    return requests.get(url, headers = {'app_id' : app_id, 'app_key' : app_key})

```
