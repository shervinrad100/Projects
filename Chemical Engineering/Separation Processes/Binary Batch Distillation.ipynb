# Binary Batch Distillation



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
    # params
# relative_volatility
a = 2.41 
# time [hr]
t = np.arange(0,10,1)
```

```python
    # Initial condition
D = 10 
W_0 = 100
x_0 = 0.5
```

```python
def batch(x_w,t): 

    # mass bal [kg]
    Wt = -D*t + W_0

    # vapour composition
    y = a * x_w / (1 + x_w * (a - 1))

    # liquid composition
    dxdt = -(D/Wt)*(y-x_w)
    
    return dxdt

x_w_t = odeint(batch,x_0,t)
```

```python
plt.figure()
plt.plot(t,x_w_t)
plt.xlabel("t [hr]")
plt.ylabel("x_w")
plt.grid()
```
