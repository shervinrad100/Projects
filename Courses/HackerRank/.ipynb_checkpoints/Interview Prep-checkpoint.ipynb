# Sock merchant



```python
def socks(n,ar):
    assert n == len(ar)
    A = {}
    pairs = 0
    ar.sort()
    print(ar)
    for i in ar:
        if i not in A:
            A[i] = 1
        else:
            A[i] += 1
        p = list(map(lambda x: x//2, list(A.values())))
    for i in p:
        pairs += i
    print(A, "each key makes the respective number of pairs", p)
    return(pairs)

if __name__ == "__main__":
    n = int(input("How many socks are there? "))
    ar = list(map(int, input("Colour code of socks? ").strip().split()))
    result = socks(n,ar)
else:
    pass
```

```python
def hourglass(ar):
    mat = []
    for j in range(4):
        for i in range(4):
            mat.append([ar[j][i],ar[j][i+1],ar[j][i+2],ar[j+1][i+1],ar[j+2][i],ar[j+2][i+1],ar[j+2][i+2]])
    A = {}
    for i in range(len(mat)):
        A[i] = mat[i]
        A[i] = sum(mat[i],)
    max = -1000
    for i in A.values():
        if i > max:
            max = i
        else:
            pass
    print(mat, A, max)

if __name__ == "__main__":
    ar = []
    for i in range(6):
        ar.append(list(map(int, input().strip().split())))
    hourglass(ar)    
```

```python

def minimumBribes(q):
    bribes = 0
    valid = True
    for i in reversed(sorted(q)):
        x = i 
        if q[-1] == x:
            q.pop(-1)
        elif len(q) >1 and q[-2] == x:
            q.pop(-2)
            bribes += 1
        elif len(q) > 2 and q[-3] == x:
            q.pop(-3)
            bribes += 2
        else:
            valid = False
    if valid:
        print(bribes)
    else: 
        print("Too Chaotic")

    
if __name__ == '__main__':
    case = int(input())
    for t_itr in range(case):
        n = int(input())
        q = list(map(int, input().rstrip().split()))
        minimumBribes(q)
```

How do you find the largest and smallest number in an unsorted integer array? 

```python
# input
import random
#ar = [1,2,6,8,5,4,3,9]
ar = []
while len(ar) < 10:
    temp = random.randint(1,100)
    if temp in ar:
        pass
    else:
        ar.append(temp)

# output
highest = ar[0]
lowest = ar[0]
for i in ar:
    if i > highest:
        highest = i
    elif i < lowest:
        lowest = i
    else:
        continue        
print(ar)
print("Highest = %i, Lowest = %i" %(highest, lowest))
```

How is an integer array sorted in place using the quicksort algorithm

```python
# input
import random
#ar = [1,2,6,8,5,4,3,9]
ar = []
while len(ar) < 10:
    temp = random.randint(1,100)
    if temp in ar:
        pass
    else:
        ar.append(temp)
        
def quicksort(arr): # Quicksort
    if len(arr) <= 1: return arr # already sorted
    pivot = arr[random.randint(0,len(arr)-1)]
    smaller, equal, larger = [],[],[]
    for x in arr:
        if x < pivot:
            smaller.append(x)
        elif x == pivot:
            equal.append(x)
        else:
            larger.append(x)
    return quicksort(smaller)+equal+quicksort(larger)

print("Unsorted:", ar)
print("Sorted:  ", quicksort(ar))
```
