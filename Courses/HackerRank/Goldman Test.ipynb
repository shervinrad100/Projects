```python
# arr = [list(map(int,input().strip().split(" "))) for _ in range(5)]
arr = [[11,12],[21,22]]
def play(arr):
    score = {"tom":0, "jerry":0}
    col = {0:[]}
    turn = "tom"
    for n in range(len(arr)):
        for m in range(len(arr[n])):
            if m not in col:
                col[m] = [arr[n][m]]
            else:
                col[m].append(arr[n][m])
    while len(col)>0:
        print("-->  ",col)
        print("turn:",turn)
        temp = 0    
        for i in col:
            if temp < max(col[i]):
                temp = max(col[i])
        for i in col:
            if temp in col[i]:
                break
        del(col[i])


        if turn == "jerry":
            score["jerry"] += temp
            turn = "tom"
        elif turn == "tom":
            score["tom"] += temp
            turn = "jerry"
        print("Choice:",temp,"\nNew:",col)
        print("score:",score)

    print("\n\tDifference",score["tom"]-score["jerry"])

play(arr)
```

```python
class Grid():
    def __init__(self, n, m):  # Creates a grid of zeros with nxm dimentions  
        self.rows = n
        self.cols = m
        self.arr = []
        for i in range(self.rows):
            self.arr.append([0 for _ in range(self.cols)])
            
    def randomArr(self): # randomly allocates 0s and 1s to the grid
        from random import choice 
        for i in range(len(self.arr)):
            for j in range(len(self.arr[i])):
                self.arr[i][j] = choice([0,1])
        return self.arr
    
    def assignArr(self, rows): # manually insert matrix
        self.arr= rows
        return self.arr
        
class Rules(Grid):
    def __init__(self, arr):
        self.arr = arr

    def liveTable(self): # presents the number of live cells around each cell (not including the current cell)
        from numpy import zeros, shape
        lives = zeros(shape(self.arr))            # create lives grid
    
        for i in range(len(self.arr)):
            for j in range(len(self.arr[i])-1):   # side by side cells
                if self.arr[i][j+1] == 1:
                    lives[i][j]+=1                         
                if self.arr[i][j] == 1:
                    lives[i][j+1] +=1
        
        for i in range(len(self.arr)-1):
            for j in range(len(self.arr[i])):     # above and below cells
                if self.arr[i+1][j] == 1:
                    lives[i][j] += 1
                if self.arr[i][j] == 1:
                    lives[i+1][j] += 1

            for j in range(len(self.arr[i])-1):   # forward diagonal
                if self.arr[i][j] == 1:
                    lives[i+1][j+1] += 1
                if self.arr[i+1][j+1] == 1:
                    lives[i][j] += 1
            
            for j in reversed(range(1,len(self.arr[i]))):  # backward diagonal 
                if self.arr[i][j] == 1:
                    lives[i+1][j-1] += 1
                if self.arr[i+1][j-1] == 1:
                    lives[i][j] += 1
            
            return lives
    
#     def edit(self, *rules): # the rule says what happens to the cells with i live neighbours
        

                
                
                
    
#     def assignRules(self, **rules): 
#             #input: rules in a dict corresponding to i live neighbours, output: grid with rules applied
#             for key,value in rules.items():
#                 print(key,value)
```

```python
def writeIn(ballot):
    ballot = sorted(ballot)
    names = {}
    for i in ballot:
        names[i] = ballot.count(i)
    temp = max(names.values())
    
    for key, val in names.items():
        if val == temp:
            winner = key
            break
    print(names)
    return winner

    
writeIn(["Alex",
"Michael",
"Harry",
"Dave",
"Michael",
"Victor",
"Harry",
"Alex",
"Mary",
"Mary"])
```
