# __Order of Growth__/Big O/Algorithmic efficiency



This is used to show the magnitude by which the computation time will increase with respect to the input dataset. 

- Rule 1: it is always the worst case scenario, so if you have to go through A and then B but A takes less time, then the big O is that of B: $O(B)$
- Rule 2: If you have constants, you drop it. So if you see two 'for' loops its not gonna be $O(2N)$. eg:

### $O(1)$
Constant growth (independent of input)

```python
A= input()
print(A)
```


### $O(N)$
Linear increase

```python
import random
n = int(input("Input dataset size ")) # O(1)
lst = [] # O(1)
for i in range(n): # n*O(1) = O(n)
    lst.append(random.randint(1,10)) # O(1)
print(len(lst)) # O(1)
```

If you have to go through two arrays, you add the time it takes to go through each $O(a+b)$. eg:

### $ O(N^2) $
Square increase because it gets multiplied by the dimension of the input.

```python
size = [] # O(1)
for i in range(len(lst)): # n*O(1)
    for j in lst: # ...*n*O(1)
        if j == i: # O(1)
            size.append("True") # O(1)
        else: # O(1)
            size.append("Flase") # O(1)
print(len(size)) # O(1)
```

### $O(2^N)$
This one (exponential) gets doubled with an additional unit of input dataset. It can be seen in recursive functions such as:

```python
def fact(n):
    if n == 1: # O(1)
        return 1 # O(1)
    else:
        return n*fact(n-1) # O(n)
    
fact(int(input().strip()))
```

If it takes $T(N)$ time units to solve the problem then $when \ N>1$:
$$T(N) = O(1) + 2T(N-1) \\
T(N) = 3O(1) + 4T(N-2) \\
T(N) = 7O(1) + 8T(N-3)\\
...\\
T(N) = (2^{N-1}-1) \times O(1) + (2^{N-1}) \times T(1)\\
T(N) = (2^N - 1)*O(1)\\
T(N) = O(2^N)$$

### $O(logN)$
Bindary search: <br />
first run, we have $n$ variables to compare, but we eliminate $n/2$ of them. This will keep going until $n==1$. So the formula is:
$$ 1: \ T(N) = T(N/2) \\
2: \ T(N) = T((N/2)/2)  \\
... \\
k: T(N) = T(N/2^k)  \\
while: \ N = 2^k , \ T(N/{2^k}) = T(2^k/2^k) = T(1) \\ 
T(N) = T(1) + log_{2}{(Nc)} \rightarrow \\
T(N) = O(log(n)) $$

```python
def binarysearch(lst, x): # Bisection search, divide and concur
    if len(lst) == 0 or (len(lst) == 1 and x not in lst): 
        print("Not Found")
        return False
    mid = lst[int(len(lst)/2)]
    if x == mid: 
        print("Found")
        return True
    if x < mid:
        return binarysearch(lst[:int(len(lst)/2)],x)
    if x > mid:
        return binarysearch(lst[int(len(lst)/2+1):],x)

for i in range(2):
    lst = list(map(int,input("array: ").strip().split(",")))
    x = int(input("search item: "))
    lst.sort()
    binarysearch(lst,x)
```
