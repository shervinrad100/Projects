# Lecture 1: Intro

Generate and Test: <br />
1. Generate a possible solution
2. Test to see if the solution fits
3. if yes that's the solution, if no, go to step 1

This can be systematic or rnadom. Some systematic algorithms are: <br />
- depth first tree search algorithm: As seen in the picture below, an arbitrary node in the graph is chosen, and you follow one branch until you get to the bottom of it before back-tracking. It's like what I do when sorting my folders when I'm taking backups. 
- Breadth first search or traversal: It goes in one level and searches all of those, and then it goes another level. ie: 1,2,7,8 then 3,6,9,12 etc
