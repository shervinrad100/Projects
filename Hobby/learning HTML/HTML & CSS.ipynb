# `<head>`
Head: Metadata usually describes data about the HTML doc and goes between `<html>` and `<body>` (what is displayed in your browser tab). It tells you about the document and what it entails and provides a setup that allows you to link your markup to relative codes later on. <br>
`<head></head>`

- `<meta>`: the element is used to specify which character set is used (`<meta charset="utf-8">`), page description, keywords for search engines (similar to tags), author (again under the same attributes), and other metadata. Metadata is used by browsers (how to display content), by search engines (keywords), and other web services. So lets summarise the attributes:

    - name, content (The content attribute gives the value associated with the http-equiv or name attribute.)
        - `<meta name="description" content="Free Web tutorials">`
        - `<meta name="keywords" content="HTML, CSS, XML, JavaScript">`
        - `<meta name="viewport" content="width=device-width, initial-scale=1.0">` <br>
            viewport element gives the browser instructions on how to control the page's dimensions and scaling. The width=device-width part sets the width of the page to follow the screen-width of the device (which will vary depending on the device). The initial-scale=1.0 part sets the initial zoom level when the page is first loaded by the browser.
    - http-equiv: The http-equiv attribute provides an HTTP header for the information/value of the content attribute.<br>
        - `<meta http-equiv="refresh" content="30">` <br>
            refreshes your page every 30 seconds <br>
        - `<meta http-equiv="content-type" content="text/html"; cahrset="utf-8">` <br>
            is the old way of saying `<meta charset="utf-8">`. However it is useful to have both because of compatibility issues with brosers    
        - `< meta http-equiv="default-style" content="preferred stylesheet">` <br>
            Specified the preferred style sheet to use.
