```python
def tictac():
    options1 = ["x", "o", "quit", "Quit", "QUIT", "exit"]
    options2 = ["y", "yes", "ye", "yh", "n", "no", "quit", "Quit", "QUIT", "exit"]
    options3 = ["quit", "Quit", "QUIT", "exit"]
    options4 = ["reset_score", "reset score", "Reset Score", "Reset score", "ResetScore", "reset Score"]
    board = ["11","12","13","21","22","23","31","32","33", "quit", "Quit", "QUIT", "exit"]
    #, "1 1", "1 2", "1 3","2 1","2 2","2 3","3 1","3 2","3 3" 
    on = True
    score = {"Player1":0, "Player2":0}        
    # playing the game 1v1
    while on:
        A = [["11","12","13"],["21","22","23"],["31","32","33"]]
        print(A[0])
        print(A[1])
        print(A[2])
        x_or_o = input("Player 1, choose either 'x' or 'o': ")
        while x_or_o not in options1:
            x_or_o = input("Choose either 'x' or 'o': ")
        player1 = x_or_o
        if player1 == "x":
            player2 = "o"
        elif player1 in options3:
            break
        else:
            player2 = "x"
        player = 1
        chosen = []
        gameover = False
        while not gameover: 
            print("To go back to main menu enter 'quit'")
            pos = input("Player %i, where do you want to place your marker (Input in format '#row #col')?: " %player).strip()
            while pos not in board or pos in chosen:
                pos = input("Occupied space or wrong input format '\n ").strip()
            if pos in options3:
                break
            else:
                pass
            chosen.append(pos)
            A[int(list(pos)[0])-1][int(list(pos)[-1])-1] = x_or_o
            print(A[0])
            print(A[1])
            print(A[2])
            # win conditions
            # diagonal
            win = []
            for i in range(3):
                win.append(A[i][i])
            if win[0] == win[1] and win[0] == win[2]:
                print("Player %s is the winner" %player)
                gameover = True
                score["Player%i" %player] += 1
            else:
                pass
            # conjugated diag
            for i in reversed(range(3)):
                win[-(i+1)] = A[i][-(i+1)]
            if win[0] == win[1] and win[0] == win[2]:
                print("Player %s is the winner" %player)
                gameover = True
                score["Player%i" %player] += 1
            else:
                pass
            # horizontal
            for i in range(3):
                win = []
                for j in range(3):
                    win.append(A[i][j])
                if win[0] == win[1] and win[0] == win[2]:
                    print("Player %s is the winner" %player)
                    gameover = True
                    score["Player%i" %player] += 1
                    break
                else:
                    continue
            # vertical
            for i in range(3):
                win = []
                for j in range(3):
                    win.append(A[j][i])
                if win[0] == win[1] and win[0] == win[2]:
                    print("Player %s is the winner" %player)
                    gameover = True
                    score["Player%i" %player] += 1
                    break
                else:
                    continue
            # Draw conditions
            if len(chosen) == 9:
                print("Draw!")
                gameover = True
            else:
                pass
            
            if gameover == True:
                break
            else:
                pass
            
            if player == 1:
                x_or_o = player2
                player = 2
            else:
                x_or_o = player1
                player = 1
        print("score:", score)
        reset_score = input("To reset the score enter 'reset_score' or to show score enter 'show'").strip()
        if reset_score in options4:
            score = {"Player1":0, "Player2":0}  
        elif reset_score == "show":
            print("score:", score)
        elif reset_score in options3:
            break
        else:
            pass
        play = input("Do you want to play again? [y/n] ")
        while play not in options2:
            play = input("Do you want to play again? [y/n] ")  
        if play == "y" or play == "yes":
            on = True
        else: 
            on = False
tictac()
```
