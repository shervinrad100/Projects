## Exercise 5.1 - Guessing Game
Create a class to play a guessing game:
At the start of the game, a random number between 0 and 50 is chosen.
The player has N attempts to guess this number.
After a guess, several things can happen:
- The player gets the number right and wins. The game terminates.
- The player gets the number wrong. The number of attempts is decremented.
  - If the number of attempts is zero, the player loses and the game terminates.
  - Otherwise, the game tells the player whether the true number is higher or lower than the guess.

You can use the 'input' built-in function to get input from the user. 

```python
import sys
```

```python

class GuessingGame():
    import random
    
    def __init__(self,n):
        self.ans = random.randint(0,10)
        self.guesses = 0
        self.n = n
        self._on = True
        
    def __main__(self):
        while self._on:
            guess = input('Enter your guess number: ')
            self.play(int(guess))
        
    def play(self, guess):
        self.guesses += 1 
        if guess == self.ans:
            print('Thats correct! terminating')
            self._on = False
        elif guess != self.ans and self.guesses == self.n:
            print('Nope. Terminating')
            self._on = False
        else:
            if guess > self.ans:
                print('Nope. Go lower. Keep guessing...')
            elif guess < self.ans:
                print('Nope. Go higher. Keep guessing...')
            
    # implement main class and run game in while loop
```

## Exercise 5.2 - Bank accounts (`*`)
- Create a class called `StandardAccount`. The class should be initialized with:
  - `balance`: the starting balance
  - `interest_rate`: the interest
- The class should implement the following:
  - `withdraw(amount)`: reduces the balance in the account by the amount plus transaction fee. Returns the amount.
  - `deposit(amount)`: adds the amount to the balance.
  - `calculate_interest()`: adds/subtracts the interest from the current balance.
- Create a class called `BadAccount`. The class should be the same as the `StandardAccount`, except that it is initialized with a transaction_fee.
  For example, `BadAccount(100, 0.01)` initializes a bank account with a starting balance of 100 and a transaction fee of 1% per transaction.
  Every time a user withdraws from the `BadAccount`, a transaction fee proportional to the amount is taken.
- Create a class called `PremierAccount`. The class should be the same as the StandardAccount, except that it is initialized with `min_amount`.
  For example, `PremierAccount(100, 5)` initializes a bank account with a starting balance of £100 and a min_amount of £5.
  When the user withdraws over `min_amount` in a transaction, they get a point. Every time they accumulate 5 points, £5 is added to the account balance.
- Implement the appropriate error checks.


```python
class StandardAccount():
    
    def __init__(self, balance, ir):
        '''initialise account with balance and interest rate
        assertion: interest rate as percent'''
        self.bal = balance
        assert ir < 1 
        self.ir = ir
        
    def withdraw(self,amt): 
        '''enter amount to be withdrawn from balance'''
        self.bal -= amt
                
    def deposit(self,amt):
        '''enter amount to be added to balance'''
        self.bal += amt
        
    def calculate_interest(self):
        '''adds interest to bal and prints new bal'''
        self.bal *= (1+self.ir)
        print('%.2f' %(self.bal))
        
        
class BadAccount(StandardAccount):
    
    def __init__(self, tran_fee,balance, ir):
        '''initialise account with transaction fee, balance and interest rate'''
        super().__init__(balance, ir) 
        self.tran_fee = tran_fee
      
    def withdraw(self, amt):
        '''enter amount to be withdrawn, bad account also charges transaction fee'''
        super().withdraw(amt)
        self.bal -= self.tran_fee*amt
    
class PremierAccount(StandardAccount):
    
    def __init__(self, minamt, balance, ir):
        '''initialise account with min transaction amount, balance and interest rate
        points start at 0'''
        super().__init__(balance, ir)
        self.minamt = minamt
        self.points = 0
        
    def withdraw(self, amt):
        '''enter amount to be withdrawn
        withdrawl larger than min_amt result in points being added
        add £5 when 5 points collected'''
        super().withdraw(amt)
        if amt > self.minamt:
            self.points += 1
        if self.points%5 == 0:
            super().deposit(5)
```
