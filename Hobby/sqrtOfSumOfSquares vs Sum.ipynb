```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sums = []
sqrtOfSums = []
xs = list(range(-100,100))
for upper in xs:
    sums.append(sum([num for num in range(-100,upper)]))
    sqrtOfSums.append(np.sqrt(sum([num**2 for num in range(-100,upper)])))
```

```python
plt.figure()
plt.plot(xs,sums, label="sums")
plt.plot(xs,sqrtOfSums, label="sqrtOfSums")
plt.legend()
plt.grid()
plt.figure()
plt.plot(xs,sqrtOfSums, label="sqrtOfSums")
plt.legend()
plt.grid()
```

```python
import random as random 
```

```python
sums = []
sqrtOfSums = []
xs = list(range(1,101))
for _ in xs:
    ys = [random.randint(1,10) for _ in range(100)]
    sums.append(sum(ys))
    sqrtOfSums.append(np.sqrt(sum([num**2 for num in ys])))
```

```python


fig, ax1 = plt.subplots()

ax1.plot(xs,sums, label="sums", color="red")
ax1.tick_params(axis='y', labelcolor="red")


ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.plot(xs,sqrtOfSums, label="sqrtOfSums", color="blue")
ax2.tick_params(axis='y', labelcolor="blue")

fig.legend()
ax2.grid()
```

- with square root of the sum of squares, the larger residuals have greater significance.
- the magnitude of the errors are larger in the sum but the trends are the same 
- sum of numbers grow exponentially faster than the sqrt of sum of squares

### larger squared numbers grow exponentially 
$$ \sqrt{(1+2+3)^2} = \sqrt{6^2} = \sqrt{36} $$
$$ \sqrt{(1^2+2^2+3^2)} = \sqrt{14}$$
$$36>14$$
$6^2$ grows much faster than $(1^2+2^2+3^2)$. So the sum is basically the summarised version of the sqrt of the squares which in essense grows faster than the square root of the sum of squares
