# Stamp Duty

Define the tax brackets 

```python
def rate(x, band=None): # defines bands and returns [rate, upper, lower, band_label]
    if (x <= 125000 and band == None) or band == 1:
        return [0,125000,0,1]
    elif (x <= 250000 and x> 125000 and band == None) or band == 2:
        return [0.02,250000,125000,2]
    elif (x <= 920000 and x > 250000 and band == None) or band == 3:
        return [0.05,920000,250000,3]
    elif (x <= 1500000 and x > 920000 and band == None) or band == 4:
        return [0.1,1500000,920000,4]
    elif (x> 1500000 and band == None) or band == 5:
        return [0.12,10**10,1500000,5]
```

Write a function that calculates the Stamp Duty Tax given a price of property

```python
def tax(P):
    main = rate(P)
    temp = rate(0,band=main[-1]-1)   
    tax = (P-main[2])*main[0]
    if main[-1] != 1:
        for i in range(1,temp[-1]+1):
            bracket = rate(0,band=i)
            tax+= bracket[0]*(bracket[1]-bracket[2])
    else:
        pass
    return tax
```

Test function against known cases:

```python
testcases = [100000,200000,500000,920000,1500000]
answers = [0,1500,15000,36000,94000]
for i in range(len(testcases)):
    if tax(testcases[i]) == answers[i]:
        print(f"test case {i}: passed")
    else:
        print(f"test case {i}: failed")
```

Now we need to find what is the optimum price at which we can pay the smallest ratio of tax to the total money spent (or maximum ratio of price to the total paid sum). 

The desired case would be to find local maxima or minima for the latter functions which shows optimality.

```python
import numpy as np
import matplotlib.pyplot as plt

price = np.arange(0,2000000,1000) # every £1,000 tax and price is re-calculated
total, stamp, stampToTotal, priceToTotal = [],[],[],[]

for i in price:
    total.append(i+tax(i))
    stamp.append(tax(i))
for i in range(len(price)):
    stampToTotal.append(stamp[i]/total[i])
    priceToTotal.append(price[i]/total[i])
    
fig, ax1 = plt.subplots()

ax1.plot(price,stamp, label="Stamp Duty")
ax1.set_xlabel("Price (£)")
ax1.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
ax1.set_ylabel("Stamp Duty (£)")
ax1.grid(axis="both")


ax2 = ax1.twinx()
ax2.plot(price,total,label="Total", color="r")
ax2.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
ax2.set_ylabel("Total (£)", color="r")
ax2.tick_params(axis='y', labelcolor="r")


plt.figure()
plt.plot(price,stampToTotal, label="Tax to Total ratio")
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.grid()
plt.legend()

plt.figure()
plt.plot(price,priceToTotal, label="Price to Total ratio")
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.grid()
plt.legend()
```
