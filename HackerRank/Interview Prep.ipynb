map() applies a function or a list of functions to inputs 

```python
def times(x):
    return x*x
def add(x):
    return x+x
funcs = [times, add]
for i in range(5):
    ans = list(map(lambda x: x(i), funcs))
    print(ans)
```

# Sock merchant



```python
def socks(n,ar):
    assert n == len(ar)
    A = {}
    pairs = 0
    ar.sort()
    print(ar)
    for i in ar:
        if i not in A:
            A[i] = 1
        else:
            A[i] += 1
        p = list(map(lambda x: x//2, list(A.values())))
    for i in p:
        pairs += i
    print(A, "each key makes the respective number of pairs", p)
    return(pairs)

if __name__ == "__main__":
    n = int(input("How many socks are there? "))
    ar = list(map(int, input("Colour code of socks? ").strip().split()))
    result = socks(n,ar)
else:
    pass
```

```python
def hourglass(ar):
    mat = []
    for j in range(4):
        for i in range(4):
            mat.append([ar[j][i],ar[j][i+1],ar[j][i+2],ar[j+1][i+1],ar[j+2][i],ar[j+2][i+1],ar[j+2][i+2]])
    A = {}
    for i in range(len(mat)):
        A[i] = mat[i]
        A[i] = sum(mat[i],)
    max = -1000
    for i in A.values():
        if i > max:
            max = i
        else:
            pass
    print(mat, A, max)

if __name__ == "__main__":
    ar = []
    for i in range(6):
        ar.append(list(map(int, input().strip().split())))
    hourglass(ar)    
```
